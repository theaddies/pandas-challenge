```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
print(school_data)
print(student_data)
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#In hindsight combinging these datasets made some of the operations below more difficult.  I should have done some on the original set.
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

```python
#this would have been easier if instead I used school_data DF for some of the calcs.
school_data_complete
```

```python
#Calculate the total number of schools
#Calculate the total number of students
total_schools = school_data_complete['school_name'].nunique()
total_students_unique = school_data_complete['student_name'].nunique()
total_students = school_data_complete['student_name'].count()
print(f"Total number of schools = {total_schools}")
#There are a lot of overlapping names and I only did that thinking it might be helpful later on.
print(f"Total number of unique student names = {total_students_unique}")
print(f"Total number of students = {total_students}")
```

```python
#Calculate the total budget
school_names = school_data_complete['school_name'].unique()
print(school_names)
total_budget = 0
for school in school_names:
    school_cluster = school_data_complete.loc[school_data_complete['school_name'] == school,:]
    #this only works because the budget is the same for each school row.  This picks the first.  Could be any row.
    budget_one_school = school_cluster['budget'].iloc[0]
    total_budget = total_budget + budget_one_school
print(f"total budget = {total_budget}")
#  OK, so that was nuts.  There must be a better way.
school_group = school_data_complete.groupby(by = 'school_name').mean()
print(f"total budget = {school_group['budget'].sum()}")
#better!
```

```python
#Calculate the average math score
#Calculate the average reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_math_score = school_data_complete['math_score'].mean()
print(f"average math score = {average_math_score}\naverage reading score = {average_reading_score}")
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
#Calculate the percentage of students with a passing reading score (70 or greater)
#Calculate the percentage of students who passed math and reading (% Overall Passing)

percent_passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70,'math_score' ].count() / total_students * 100
percent_passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, 'reading_score'].count() / total_students * 100
percent_passing_reading_math = school_data_complete.loc[((school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)), 'reading_score'].count() / total_students * 100
print(f"percent passing math = {percent_passing_math}")
print(f"percent passing reading = {percent_passing_reading}")
print(f"percent passing reading and math = {percent_passing_reading_math}")
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting

summary_columns = ['Total number of schools','Total number of students','total budget','average math score','average reading score','percent passing math','percent passing reading','percent passing reading & math']
data_values = np.array([[total_schools,total_students,total_budget,average_math_score,average_reading_score,percent_passing_math,percent_passing_reading,percent_passing_reading_math]])
schools_summary_df = pd.DataFrame(data = data_values, columns = summary_columns)
schools_summary_df['Total number of schools'] = schools_summary_df['Total number of schools'].map("{:^.0f}".format)
schools_summary_df['Total number of students'] = schools_summary_df['Total number of students'].map("{:^,.0f}".format)
schools_summary_df['total budget'] = schools_summary_df['total budget'].map("${:,.0f}".format)
schools_summary_df['average math score'] = schools_summary_df['average math score'].map("{:.0f}".format)
schools_summary_df['average reading score'] = schools_summary_df['average reading score'].map("{:^.0f}".format)
schools_summary_df['percent passing math'] = schools_summary_df['percent passing math'].map("{:.1f}%".format)
schools_summary_df['percent passing reading'] = schools_summary_df['percent passing reading'].map("{:.1f}%".format)
schools_summary_df['percent passing reading & math'] = schools_summary_df['percent passing reading & math'].map("{:.1f}%".format)
schools_summary_df
```

```python
#School Name
#School Type
#Total Students
#Total School Budget
#Per Student Budget
#Average Math Score
#Average Reading Score

school_group_df = school_data_complete.groupby(by = ['school_name', 'type'], as_index = False).mean()
#school_group_df = school_group_df.drop(columns = ["Student ID","School ID"])
school_group_df = school_group_df.rename(columns = {'reading_score':'average reading score','math_score':'average math score','size':'number of students'})
school_group_df['per student budget'] = school_group_df['budget'] / school_group_df['number of students']
#school_group_df = school_group_df.reindex(school_data_complete['school_name'])
school_group_df = pd.DataFrame(school_group_df)
print(school_group_df.dtypes)
print(school_group_df)
type(school_group_df)
print(school_group_df.columns)
#if you want to have the normal index just take out this line.
school_group_df.set_index('school_name', inplace = True)
school_group_df
```

```python
#% Passing Math
#% Passing Reading
#% Overall Passing (The percentage of students that passed math and reading.)

#dang.  finally.  I needed to make the count a dictionary as then I can add it to the main df with no issues.
student_count = {}
for school in school_group_df.index:
    school_true = (school_data_complete['school_name'] == school)
    reading_score_pass = (school_data_complete['reading_score'] >= 70)
    math_score_pass = (school_data_complete['math_score'] >= 70)
    math_reading_score_pass = (reading_score_pass & math_score_pass)
    student_count[school] = [school_data_complete.loc[(school_true),'school_name'].count()]
    passing_reading_students = school_data_complete.loc[(school_true & reading_score_pass),'school_name'].count()
    passing_math_students = school_data_complete.loc[(school_true & math_score_pass),'school_name'].count()
    passing_math_reading_students = school_data_complete.loc[(school_true & math_reading_score_pass),'school_name'].count()
    student_count[school].append(passing_reading_students)
    student_count[school].append(passing_math_students)
    student_count[school].append(passing_math_reading_students)
    #math_passing_count = school_data_complete.loc[((school_data_complete['school_name'] == school) & (school_data_complete['reading_score'] >= 70)].count()
    #school_group_df['math_passing_count'] = school_data_complete.loc[(school_data_complete['school_name'] == school) & (school_data_complete['reading_score'] >= 70)].count()
    #print(math_passing_count)
    #reading_passing_count[] =
    #reading_math_passing_count[] 
print(student_count)
```

```python
#now lets write this dictionary back into the main school dictionary
for school in school_group_df.index:
    #print(school)
    #print(student_count[school][0])
    ###print(student_count[school][1])
    #print(school_group_df.iloc[0,0])
    #print(school_group_df.loc[school,:])
    #print(school_group_df.index)
    #print(school_group_df[school]['average math score'])
    percent_reading_pass = student_count[school][1] / student_count[school][0] * 100
    percent_math_pass = student_count[school][2] / student_count[school][0] * 100
    percent_math_reading_pass = student_count[school][3] / student_count[school][0] * 100
    school_group_df.loc[school,'percent reading pass'] = percent_reading_pass
    school_group_df.loc[school, 'percent math pass'] = percent_math_pass
    school_group_df.loc[school, 'percent reading & math pass'] = percent_math_reading_pass
#the new data frame has all data that the old dataframe had plus the new information
print(school_group_df)
print(school_data)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school

math_9th_mean = pd.Series([], dtype = float)
math_10th_mean = pd.Series([], dtype = float)
math_11th_mean = pd.Series([], dtype = float)
math_12th_mean = pd.Series([], dtype = float)
for school in school_group_df.index:
    school_true = (school_data_complete['school_name'] == school)
    grade_9th = school_data_complete['grade'] == '9th'
    grade_10th = school_data_complete['grade'] == '10th'
    grade_11th = school_data_complete['grade'] == '11th'
    grade_12th = school_data_complete['grade'] == '12th'
    #you must set this value to a pandas series in order to append it to the series values.
    #must reindex the pandas series
    math_9th = pd.Series(school_data_complete.loc[(school_true & grade_9th),'math_score'].mean())
    math_9th.index = [school]
    math_9th_mean = math_9th_mean.append(pd.Series(math_9th))
    math_10th = pd.Series(school_data_complete.loc[(school_true & grade_10th),'math_score'].mean())
    math_10th.index = [school]
    math_10th_mean = math_10th_mean.append(pd.Series(math_10th))
    math_11th = pd.Series(school_data_complete.loc[(school_true & grade_11th),'math_score'].mean())
    math_11th.index = [school]
    math_11th_mean = math_11th_mean.append(pd.Series(math_11th))
    math_12th = pd.Series(school_data_complete.loc[(school_true & grade_12th),'math_score'].mean())
    math_12th.index = [school]
    math_12th_mean = math_12th_mean.append(pd.Series(math_12th))

    
print("math_9th_mean")    
print(math_9th_mean)
print("math_10th_mean")    
print(math_10th_mean)
print("math_11th_mean")    
print(math_11th_mean)
print("math_12th_mean")    
print(math_12th_mean)
```

```python
#Combine the series into a dataframe
#Optional: give the displayed data cleaner formatting
school_grade_math_df = pd.DataFrame({'9th':math_9th_mean, '10th' : math_10th_mean, '11th' : math_11th_mean, '12th' : math_12th_mean})
#school_grade_math_df.transpose()
school_grade_math_df
school_grade_math_df['9th'] = school_grade_math_df['9th'].map("{:.0f}".format)
school_grade_math_df['10th'] = school_grade_math_df['10th'].map("{:.0f}".format)
school_grade_math_df['11th'] = school_grade_math_df['11th'].map("{:.0f}".format)
school_grade_math_df['12th'] = school_grade_math_df['12th'].map("{:.0f}".format)
print(school_grade_math_df)
```
